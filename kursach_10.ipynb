```python
"""pervoe zadanie - bystraya sortirovka Hoara"""

import random 
Arr = list(map(int, input().split()))
def sortirovka(arr):
    if len(arr) <= 1:
        return arr
    else:
        bar = random.choice(arr)
        left_part = [elem for elem in arr if elem < bar]
        ravnye = [bar] * arr.count(bar)
        right_part = [elem for elem in arr if elem > bar] 
        return sortirovka(left_part) + ravnye + sortirovka(right_part)
print (sortirovka(Arr))
```

```python
"""vtoroe zsdanie - dvusvyazhyy spisok"""

class point():
    def __init__(self, data, prev, next):
        self.data = data
        self.prev = prev
        self.next = next

class linkedlist():
    def __init__(self):
        self.head = None
        self.tail = None
        
    def add_point(self, data):
        new_point = point(data, None, None)
        if self.head is None:
            self.head = self.tail = new_point
        else:
            new_point.prev = self.tail
            new_point.next = None
            self.tail.next = new_point
            self.tail = new_point

    def delete_point(self, num):
        counter = 0
        current = self.head
        while counter <= num:
            if counter == num:
                if current.prev is None:
                    self.head = current.next
                    current.data = None
                else:
                    current.prev.next = current.next
                    current.next.prev = current.prev
            counter += 1
            current = current.next
            
    def vyvod(self):
        if self.head is None:
            print(None)
        else:
            current_point = self.head
            while current_point is not None:
                if current_point.next is None:
                    print(current_point.data, end=" ")
                    break
                else:
                    print(current_point.data, end=" ")
                    current_point = current_point.next
                    
    def getitem(self, num):
        counter = 0
        current = self.head
        while counter < num:
            counter += 1
            current = current.next
        return current.data
        
Arr = linkedlist()

Arr.add_point(input())
Arr.add_point(56)
Arr.add_point(57)
print (Arr.vyvod())

Arr.delete_point(int(input()))

print (Arr.vyvod())
print (Arr.getitem(int(input())))
```

```python
"""trit'e zadanie - okaymlyayuschie minory dlya ranga matrici"""

def minor(M,n,k):
    Min=[[0] * (n-1) for i in range(n-1)]
    for i in range(1,n):
        for j in range(k):
            Min[i-1][j]=M[i][j]
        for j in range(k+1,n):
            Min[i-1][j-1]=M[i][j]
    return Min

def det(M,n):
    D=0
    if n==1:
        return M[0][0]
    else:
        for i in range(n):
            D+=((-1)**i)*M[0][i]*det(minor(M,n,i),n-1)
    return D

def mat(M,K,L):
    n=len(K)
    Min=[[0] * n for i in range(n)]
    f=0
    s=0
    for i in K:
        for j in L:
            Min[f][s]=M[i][j]
            s+=1
        s=0
        f+=1
    return Min

def ok(M,n,m):
    i=0
    K=[]
    L=[]
    j=0
    if m<n:
        r=m
    else: r=n
    while i<n and (j==m or M[i][j]==0):
        j=0
        while j<m and M[i][j]==0:
            j+=1
        if j==m:
            i+=1
    if i==n:
        return 0
    L.append(j)
    K.append(i)
    i=1
    j=0
    f=0
    while i<r:
        if j in K:
            j+=1
        elif f in L:
            f+=1
        elif f==m:
            f=0
            j+=1
        elif j==n:
            return i
        else:
            K.append(j)
            L.append(f)
            if f<m and j<n and det(mat(M,K,L),len(K))!=0:
                i+=1
                j=0
                f=0
            elif f==m or j==n:
                return i
            else:
                K.pop()
                L.pop()
                f+=1
    return i
  
print("Enter matrix size:")
n, m = list(map(int, input().split()))
print("Enter matrix:")
M = [list(map(int, input().split())) for i in range(n)]
print(ok(M,n,m))
```
